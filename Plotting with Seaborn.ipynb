### Tips Data Analysis

Analyzing the famous tips data set to get more familiar with seaborn.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
#setting color palette, plotting the tip distribution
sns.set_palette("Spectral")
ax = sns.violinplot(x="sex", y="tip", inner='quartile', data=tips)
ax.set_title('Distribution of tips', fontsize=16);
```

```python
sns.violinplot(data=tips,x='sex', y='tip', hue='time',split=True)
```

```python
sns.catplot(data=tips, x='sex', y='total_bill',  hue='time',  kind='violin', split=True)
```

```python
sns.kdeplot(data=tips, x='tip')
```

```python
sns.kdeplot(data=tips, x="total_bill", hue="time")
```

```python
df= sns.load_dataset("taxis")
```

```python
#adding drivetime as a new column
df['drivetime']=((df['dropoff']-df['pickup']).dt.seconds)/60
df
```

```python
df=df.drop("total", axis='columns')
df['total']=df.fare+df.tip+df.tolls
df
```

```python
sns.heatmap(df.corr(numeric_only=True),annot=True)
```

```python
df.describe()
```

```python
sns.violinplot(data=df,x='passengers', y='tip', hue='color', split=True)
```

```python
sns.violinplot(data=df,x='passengers', y='drivetime', hue='color', split=True)
```

```python
df[['tip','color']].boxplot(by='color')
```
